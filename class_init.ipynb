```python
class Test:
    def __init__(self,age,name,family) -> None:

        self.name = name
        self.age = age
        self.family = family
        # مقادیری که باید درون کلاس ساخته شوند را می سازد
        self.CreatFirstData() 
        # تابع های که باید بلافاصله با صدا زدن کلاس اجرا شوند را صدا میزند
        self.CreatFirstFun()


    def CreatFirstData(self):
        self.describ =  f'Im {self.name} {self.family} and i have { self.age} years old. '
        self.country = 'Iran'
        self.city    = 'Tehran'
        self.street  = 'ValiAsr'


    def CreatFirstFun(self):
        # لیست تابع های که در جا اجرا می شوند
        self.printA()
        self.printB()
        self.printC()
        self.printD()

    def printA(self):
        print(self.country)
    def printB(self):
        print(self.city)
    def printC(self):
        print(self.street)
    def printD(self):
        print(self.describ)  

```
